```python

def check_prerequisites_met(course, finished_courses_set, catalog_df):
    """
    Check if all prerequisites for a course are met
    
    Returns: (bool: prerequisites_met, list: missing_prerequisites)
    """
    course_row = catalog_df[catalog_df['Course Num'] == course]
    
    if course_row.empty:
        return False, ["Course not found in catalog"]
    
    prereqs = course_row['Pre_Requisites'].values
    
    # If no prerequisites, course is available
    if pd.isna(prereqs) or prereqs == '' or str(prereqs).strip() == '':
        return True, []
    
    missing = []
    
    # Parse prerequisites (handle "or" and "and" cases)
    prereq_list = str(prereqs).replace(' or ', ',').replace(' and ', ',').split(',')
    prereq_list = [p.strip() for p in prereq_list if str(p).strip()]
    
    for prereq in prereq_list:
        if prereq not in finished_courses_set:
            missing.append(prereq)
    
    return len(missing) == 0, missing
def get_available_courses(finished_courses_list, catalog_df):
    """
    Get all courses that can be taken based on completed prerequisites
    """
    finished_set = set(finished_courses_list)
    available_courses = []
    
    for idx, row in catalog_df.iterrows():
        course_num = row['Course Num']
        
        # Skip if already completed
        if course_num in finished_set:
            continue
        
        # Check if prerequisites are met
        prereqs_met, missing = check_prerequisites_met(course_num, finished_set, catalog_df)
        
        if prereqs_met:
            available_courses.append({
                'Course Num': course_num,
                'Course Name': row['Course Name'],
                'Pre_Requisites': row['Pre_Requisites'] if pd.notna(row['Pre_Requisites']) else 'None',
                'Description': row['Description']
            })
    
    return pd.DataFrame(available_courses)


def get_blocked_courses_with_missing_prereqs(finished_courses_list, catalog_df):
    """
    Get courses that cannot be taken yet and show what's missing
    """
    finished_set = set(finished_courses_list)
    blocked_courses = []
    
    for idx, row in catalog_df.iterrows():
        course_num = row['Course Num']
        
        # Skip if already completed
        if course_num in finished_set:
            continue
        
        # Check prerequisites
        prereqs_met, missing = check_prerequisites_met(course_num, finished_set, catalog_df)
        
        if not prereqs_met and missing:
            blocked_courses.append({
                'Course Num': course_num,
                'Course Name': row['Course Name'],
                'Missing Prerequisites': ', '.join(missing),
                'All Prerequisites': row['Pre_Requisites']
            })
    
    return pd.DataFrame(blocked_courses)


def create_user_friendly_report(finished_courses_list, catalog_df):
    """
    Create a user-friendly report showing course availability status
    """
    print("=" * 80)
    print("COURSE AVAILABILITY REPORT")
    print("=" * 80)
    
    # Get completed courses count
    print(f"\n✓ Completed Courses: {len(finished_courses_list)}")
    
    # Get available courses
    available = get_available_courses(finished_courses_list, catalog_df)
    print(f"\n✓ Courses Available to Take NOW: {len(available)}")
    
    if len(available) > 0:
        print("\nYou can register for these courses:")
        print("-" * 80)
        for idx, row in available.head(80).iterrows():
            prereq_text = row['Pre_Requisites'] if row['Pre_Requisites'] != 'None' else 'No prerequisites'
            print(f"  • {row['Course Num']}: {row['Course Name']}")
            print(f"    Prerequisites: {prereq_text}")
            print()
    
    # Get blocked courses
    blocked = get_blocked_courses_with_missing_prereqs(finished_courses_list, catalog_df)
    print(f"\n✗ Courses Still BLOCKED: {len(blocked)}")
    
    if len(blocked) > 0:
        print("\nYou cannot take these courses yet (showing first 10):")
        print("-" * 80)
        for idx, row in blocked.head(80).iterrows():
            print(f"  • {row['Course Num']}: {row['Course Name']}")
            print(f"    MISSING: {row['Missing Prerequisites']}")
            print()
    
    return available, blocked
```

```python
def format_prerequisites_with_or_logic(prereqs_set, description_text):
    """Extract and preserve OR relationships from course description."""
    import re
    
    if not prereqs_set:
        return ""
    
    # Find "X #### or Y ####" patterns in description
    or_pattern = r'([A-Z]{2,4}\s+\d{4})\s+or\s+([A-Z]{2,4}\s+\d{4})'
    or_matches = re.findall(or_pattern, description_text, re.IGNORECASE)
    
    or_groups = []
    remaining = set(prereqs_set)
    
    for match in or_matches:
        course1 = ' '.join(match[0].strip().split())
        course2 = ' '.join(match[1].strip().split())
        
        if course1 in remaining and course2 in remaining:
            or_groups.append(f"{course1}|{course2}")
            remaining.discard(course1)
            remaining.discard(course2)
    
    all_parts = list(remaining) + or_groups
    return ', '.join(all_parts)

def check_prerequisites_met_with_or(course, finished_courses_set, catalog_df):
    """Check if prerequisites are met, handling OR logic with | separator."""
    course_row = catalog_df[catalog_df['Course Num'] == course]
    
    if course_row.empty:
        return False, ["Course not found in catalog"]
    
    prereqs = course_row['Pre_Requisites'].values[0]
    
    if pd.isna(prereqs) or prereqs == '' or prereqs.strip() == '':
        return True, []
    
    and_groups = [g.strip() for g in prereqs.split(',') if g.strip()]
    missing_groups = []
    
    for group in and_groups:
        if '|' in group:
            # OR group - need at least ONE
            or_courses = [c.strip() for c in group.split('|')]
            has_any = any(course in finished_courses_set for course in or_courses)
            if not has_any:
                missing_groups.append(f"({' or '.join(or_courses)})")
        else:
            # Single required course
            if group not in finished_courses_set:
                missing_groups.append(group)
    
    return len(missing_groups) == 0, missing_groups


def get_available_courses_with_or(finished_courses_list, catalog_df):
    """Get all courses that can be taken, handling OR prerequisites."""
    finished_set = set(finished_courses_list)
    available_courses = []
    
    for idx, row in catalog_df.iterrows():
        course_num = row['Course Num']
        if course_num in finished_set:
            continue
        
        prereqs_met, missing = check_prerequisites_met_with_or(course_num, finished_set, catalog_df)
        
        if prereqs_met:
            available_courses.append({
                'Course Num': course_num,
                'Course Name': row['Course Name'],
                'Pre_Requisites': row['Pre_Requisites'] if pd.notna(row['Pre_Requisites']) else 'None'
            })
    
    return pd.DataFrame(available_courses)

print("✓ OR-aware functions loaded!")
```
